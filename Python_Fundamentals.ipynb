Structure 1:for i in range(0, 7):user_input = int(input("Enter the number:")) if condition...-------asks the user for input seven times in a row

```python
guess_number=98

for i in range(0,7):
  user_input=int(input("Enter the number:"))

  if user_input==guess_number:
    print("Congrats! You have guessed the number")

  else:
    print("Your guess is wrong")
```

**For:** When the number of iterations is defined
**While**: When the number of iterations is not defined


```python
##Level 3:Multiple conditions with formatting
```

If the number is negative or above 100- Invalid number,Enter the numberbetween 1-100
If the user input number is > guessed number, Enter the Number is higher
If the user input number is < guessed number, Enter the Number is lower

break-used when the condition is successful to break the loop and stop the code running

```python
guess_number=98

for i in range(0,7):
  user_input=int(input("Enter the number:"))

  if user_input <0 or user_input > 100:
    print("Invalid number, Enter the number between 1-100")

  elif user_input > guess_number:
    print("Number is higher")

  elif user_input < guess_number:
    print("Number is lower")

  else:
    print(f"Congrats! You have guessed the number {guess_number} in {i} trial")
    break
```

```python
category = "programming languages"

clues = [
    "Python",
    "Java",
    "C++",
    "JavaScript",
    "Go"
]

for clue in clues:

    print(f"\nClue: {clue}")

    guess = input(
        "Guess the category (or press Enter to continue): "
    ).lower().strip()

    if guess == category:
        print("🏆 You got it!")
        break

else:
    print(f"\nAnswer: {category}")
```
